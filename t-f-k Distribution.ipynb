```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
```

```python
samplerate, data = wavfile.read('audio.wav')
```

```python
length = data.shape[0] / samplerate
print(length)
```

```python
# temp = np.asarray([])
temp = []
samplerate = int(samplerate)
for i in range(300):
#     print((data[samplerate*i:samplerate*(i+1)]))
    temp.append(sum(data[samplerate*i:samplerate*(i+1),1])/int(samplerate))
data = np.asarray(temp)
```

```python
# data = data[:len(data)//100]
data -= min(data)
data -= sum(data)/len(data)
# print(len(data))
# print(data-min(data))
data = np.asarray([i if ("0.00" in str(i)) else 0.0 for i in data])
data = abs(data)
data/=sum(data)
print(data)
```

```python
time = np.linspace(0., length, data.shape[0])
plt.figure(1)
plt.plot(time, data[:], label="Right channel")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.show()
```
